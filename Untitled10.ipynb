```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/india_job_market_dataset.csv")
df
```

```python
sns.countplot(x="Job Title",data=df,hue="Job Title",palette="viridis")
plt.xlabel("Job Title", fontsize = 15)
plt.ylabel("Count", fontsize = 15)
plt.title("Count of Roles by Job Titles", fontsize = 17, color = "black")
plt.show()
```

```python
gb = df.groupby("Job Type").agg({"Job ID":"count"}) #to create summary like pivot table
plt.pie(gb['Job ID'], labels = gb.index, autopct = "%1.2f%%") #to create a pic chart in matplotlib plt.show()
```

```python
plt.figure(figsize = (10,4))
sns.countplot(x="Job Title",data=df,hue="Salary Range",palette="GnBu" )
plt.xticks(rotation = 90)
plt.xlabel("Job Title", fontsize = 15)
plt.ylabel("Count", fontsize = 15)
plt.title("Count of Roles by Job Types", fontsize = 17, color = "black")
plt.show()
```
